```
# 1- viết function call_vectors
def call_vectors(op,u,v):
  result = None
  if op == '+':
    result = u + v
  elif op == '-':
    result = u - v
  elif op == '*':
    result = u * v
  elif op == '/':
    result = u / v
  elif op == 'dot':
    result = u.dot(v)
  else:
    result = None
  return result
    
```

```
u = np.array([1,2,3,4,5])
v = np.array([1,2,3,4,5])
call_vectors('dot',u,v)
```

```
call_vectors('#',u,v)
```

```
# 2- viết function call_vectors_norm
def call_vectors_norm(t,v):
  result = None
  if t == 'L1':
    result = norm(v,1)
  elif t == 'L2':
    result = norm(v)
  elif t == 'max':
    result = norm(v,math.inf)
  else:
    result = None
  return result
```

```
v = np.array([1,2,3,4,5,6,7])
call_vectors_norm('a',v)
```

```
# 3- viết function call_matrices
def call_matrices(op,m1,m2):
  result = None
  if op == '+':
    result = m1 + m2
  elif op == '-':
    result = m1 - m2
  elif op == '*':
    result = m1 * m2
  elif op == '/':
    result = m1 / m2
  else:
    result = None
  return result
```

```
m1 = np.array([[1,3,2,4],[4,3,6,8],[5,6,3,1]])
m2 = np.array([[2,4,5,1],[5,1,2,3],[1,8,1,9]])
call_matrices('+',m1,m2)
call_matrices('-',m1,m2)
```

```
# 6 Viết function cho phép người dùng nhập vào 1 ma trận (m*n)
def creat_matrix(m,n):
  lst = []
  for i in range(m):
    lst_sub = []
    for j in range(n):
      s = "M[" + str(i+1) + "," + str(j+1) + "]:"
      x = eval(input(s))
      lst_sub.append(x)
    lst.append(lst_sub)
  return np.array(lst)
```

```
# 7 Viết function tạo ma trận ngẫu nhiên
import random
def creat_matrix_random(m,n,start,end):
  lst = []
  for i in range(m):
    lst_sub = []
    for j in range(n):
      x = random.randint(start,end+1)
      lst_sub.append(x)
    lst.append(lst_sub)
  return np.array(lst)


```

```
def creat_vectors_random(n,start,end):
  lst = []
  for i in range(n):
    x = random.randint(start,end+1)
    lst.append(x)
  return np.array(lst)
```
